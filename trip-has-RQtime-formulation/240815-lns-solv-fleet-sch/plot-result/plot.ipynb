```python
import matplotlib.pyplot as plt

# Selected text data
data = """
Number of delivered people	runtime
51	0
74	35
86	91
94	178
104	61
120	108
132	64
146	1321
"""
data2 = """
Number of delivered people,runtime 
84,0
120,43
138,294
152,206
168,92
186,66
204,143
228,299
250,731
276,2741
"""

# Parsing the data
delivered_people = []
runtime = []
delivered_people2 = []
runtime2 = []
# Split the data into lines and iterate over them, skipping the header
for line in data.strip().split('\n')[1:]:
    people, time = line.split('\t')
    delivered_people.append(int(people))
    runtime.append(int(time))

for line in data2.strip().split('\n')[1:]:
    people2, time2 = line.split(',')
    delivered_people2.append(int(people2))
    runtime2.append(int(time2))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(delivered_people, runtime, marker='o', linestyle='-', label='10 Agents')  # Line plot
plt.plot(delivered_people2, runtime2, marker='x', linestyle='-', label='30 Agents')  # Plotting for data2
# For a scatter plot, use: plt.scatter(delivered_people, runtime)

plt.title('Weighted Edge-time Assignment; Max Running Time first round 60; timestep=11')
plt.xlabel('Number of Delivered People')
plt.ylabel('Runtime')
plt.legend()
# plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import defaultdict

# Read the file
file_path = '/Users/duong/Work-Documents/NMSU/NASA_ULI_2022/git-public/evtol-scheduling/trip-has-RQtime-formulation/240815-lns-solv-fleet-sch/instance/mer_lmp.lp'
with open(file_path, 'r') as file:
    data = file.readlines()

# Parse the data
mer_data = defaultdict(list)
lmp_data = defaultdict(list)

for line in data:
    if line.startswith('mer'):
        parts = line.strip('mer().\n').split(',')
        vertiport_name = parts[0].strip()
        time = int(parts[1].strip())
        value = int(parts[2].strip())
        mer_data[vertiport_name].append((time, value))
    elif line.startswith('lmp'):
        parts = line.strip('lmp().\n').split(',')
        vertiport_name = parts[0].strip()
        time = int(parts[1].strip())
        value = int(parts[2].strip())
        lmp_data[vertiport_name].append((time, value))

# Plot the data
for vertiport_name, points in mer_data.items():
    times, values = zip(*points)
    plt.figure(figsize=(10, 6))
    plt.plot(times, values, label=f'mer - {vertiport_name}', marker='')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.title(f'Plot of mer Data for {vertiport_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

for vertiport_name, points in lmp_data.items():
    times, values = zip(*points)
    plt.figure(figsize=(10, 6))
    plt.plot(times, values, label=f'lmp - {vertiport_name}', marker='')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.title(f'Plot of lmp Data for {vertiport_name}')
    plt.legend()
    plt.grid(True)
    plt.show()
```
